Pregunta 6 - Cuadro de mando de acciones e ingresos de GameStop - 2 Puntos

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data['Date'], data['Close'], label='Precio de cierre', color='blue')
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel('Precio de cierre (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Obtener los datos históricos de gme
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Restablecer el índice
gme_data.reset_index(inplace=True)

# Llamar a la función make_graph con título personalizado
make_graph(gme_data, "Evolución del precio de las acciones de gme")
```
